# 0. Utils folder path setup for imports

Before we start we need to add the utils folder to the system path so that we can import the functions defined there.

```python
import sys
from pathlib import Path as pth
import gc

# Add the utils folder to the system path
project_root = pth.cwd().parent # Should go from ML_Project/notebooks to ML_Project

sys.path.insert(0,str(project_root))

print("Utils folder added to system path for imports.")
```

# 1. Importing Libraries

In this cell, we will import all the necessary libraries and functions for model evaluation.

```python
from utils.evaluation import evaluate_model
import pickle as plk
import gc
import IPython
from utils.visualization import plot_model_comparison
```

## 2. Loading Trained Models from the previous notebook

## 2.1 Loading SVD model trained on 1M dataset and evaluating it

In this cell, we will load the SVD model trained on the 1M dataset from disk and evaluate its performance using predefined evaluation functions.

```python
# Defining the path to load the trained SVD model for 1M dataset
svd_1m_path = project_root / 'models' / 'svd_1m.pkl'

# Loading the trained SVD model for 1M dataset
with open(svd_1m_path, 'rb') as f:
    svd_1m = plk.load(f)

# Loading the test dataset for 1M dataset
test_1m_path = project_root / 'data' / 'prepared-1m' / "test_1m.pkl"
with open(test_1m_path, 'rb') as f:
    test_1m = plk.load(f)

# Evaluating the SVD model on the 1M test dataset
svd_1m_evaluation = evaluate_model(svd_1m, test_1m)
```

#### 2.1.1 Cleanup after evaluating SVD on 1M dataset

In this cell, we will delete the loaded SVD model for the 1M dataset from memory and run garbage collection to free up memory.

```python
# Deleting the loaded SVD model
del svd_1m

# Running garbage collection to free up memory
gc.collect()
```
